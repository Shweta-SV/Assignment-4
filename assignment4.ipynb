```python
#This code verifies the value of k 
#The problem was to find the value of k when a given equation represents a pair of straight lines.
#The value of k was found as- k=-3,k=-7
import numpy as np
#Taking k=-3
a = np.array([[6,-1.5,2],[-1.5,-3,2.5],[2,2.5,-2]])
#Discriminant 
det1 = np.linalg.det(a)
print(det1)
if det1 == 0:
    print('Value of k is correct')
else:
    print('Value of k is incorrect')
    
#Taking k=-7
b = np.array([[6,-3.5,2],[-3.5,-3,2.5],[2,2.5,-2]])
#Discriminant 
det2 = np.linalg.det(b)
if np.ceil(det2) == 0:
    print('Value of k is correct')
else:
    print('Value of k is incorrect')
    
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-5,5,100)
y_1 = -2*x+2/3
y_2 = x+1
plt.plot(x,y_1,'-r',label='-2x+2/3')
plt.plot(x,y_2,'-b',label='x+1')
plt.title('Graph of a pair of straight lines when k=-3')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper right')
plt.grid()
plt.show()    

import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-5,5,100)
y_1 = -3*x+1
y_2 = (0.667)*x+0.667
plt.plot(x,y_1,'-r',label='-3x+1')
plt.plot(x,y_2,'-b',label='(2/3)x+2/3')
plt.title('Graph of a pair of straight lines when k=-7')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper right')
plt.grid()
plt.show()    
```
